```python
import numpy as np 
import matplotlib.pyplot as plt

punto1=[0,0]
punto2=[3,0]
punto3=[3,4]
x_coords, y_coords = zip(punto1, punto2, punto3, punto1)
plt.scatter(x_coords, y_coords, color = 'red' )
plt.plot(x_coords, y_coords, color = 'green')
plt.grid(alpha = 0.3)
plt.savefig('/home/marco/Documentos/docencia/seminario_didactica_aritmetica/images/triangulo_rectangulo_semicircunferencias.png')
plt.axis('equal')
# trazado de primera semicircunferencia
plt.show()
```
