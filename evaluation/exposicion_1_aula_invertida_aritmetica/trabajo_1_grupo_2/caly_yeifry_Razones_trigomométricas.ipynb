# **¡Bienvenidos a la clase de razones trigonométricas en aula invertida!**

Somos el segundo equipo, conformado por Yeifry Uribe y Estefania Caly, y estaremos implementando la estrategia educativa conocida como "aula invertida" (El aula invertida es un modelo de enseñanza en el que los estudiantes estudian el contenido previamente fuera del aula, a través de videos, lecturas u otros recursos educativos, y luego, en clase, se dedican a la aplicación práctica de ese contenido con la guía del profesor. De esta manera, el tiempo en clase se utiliza de forma más efectiva para resolver dudas, trabajar en proyectos colaborativos y aplicar el conocimiento adquirido.), lo que significa que tendrán que ver un video previamente y luego trabajaremos en ejercicios prácticos en clase.

Para nuestra próxima sesión en persona, trabajaremos en ejercicios prácticos para aplicar lo que han aprendido en el video. Dividiremos la clase en grupos y cada grupo trabajará en un conjunto de problemas relacionados con razones trigonométricas. Se les dará acceso a herramientas en línea como calculadoras y simuladores para ayudarles en su trabajo. Además, se les solicita llevar un computador o en su defecto prestar un equipo en la universidad (esto se debe hacer antes del inicio de clase).

A medida que trabajen en los problemas, estaremos disponibles para responder a cualquier pregunta que puedan tener y para proporcionar orientación adicional si es necesario. También les animo a que trabajen juntos en sus grupos y se ayuden mutuamente.

Finalmente, terminaremos la clase con una breve discusión en grupo para compartir y discutir los enfoques utilizados para resolver los problemas, así como para abordar cualquier pregunta o confusión que haya surgido durante la actividad.

Les animamos a que tomen notas y consulten cualquier duda a través del correo electrónico.

### Prerrequisitos básicos que los estudiantes deben haber aprendido previamente:

* Conocimiento de ángulos: Los estudiantes deben saber cómo medir y nombrar ángulos, así como conocer las unidades de medida utilizadas (grados y radianes).

* Trigonometría básica: Los estudiantes deben saber qué es una función trigonométrica (seno, coseno y tangente) y cómo se relacionan con un triángulo rectángulo. Deben estar familiarizados con el teorema de Pitágoras y cómo se puede usar para calcular las longitudes de los lados de un triángulo rectángulo.

* Conceptos básicos de geometría: Los estudiantes deben conocer los conceptos básicos de geometría, como la definición de un triángulo, las propiedades de los ángulos y los tipos de triángulos.

* Álgebra básica: Los estudiantes deben estar familiarizados con la resolución de ecuaciones simples y cómo se pueden aplicar a problemas relacionados con la trigonometría.

Recuerden que, como estudiantes universitarios, es importante que desarrollen habilidades de aprendizaje autónomo y que utilicen la tecnología para ampliar su conocimiento. ¡Estamos emocionados de ver el progreso que todos haremos en este tema!
