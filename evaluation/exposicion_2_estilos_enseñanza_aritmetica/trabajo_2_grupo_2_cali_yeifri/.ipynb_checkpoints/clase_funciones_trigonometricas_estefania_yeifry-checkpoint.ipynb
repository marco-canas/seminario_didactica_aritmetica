## ¬øque son funciones trigonometricas?



> Es el comportamiento de la razones trigonometricas para todo los valores de $x$. Son las funciones cuyo argumento, o variable independiente, es un √°ngulo.

Definicion matematica



> Dada una circunferencia de radio $r$, al trazar dos ejes perpendiculares entre s√≠ que pasan por el origen del c√≠rculo, tendremos un sistema de ejes coordenados XY. Para calcular las diferentes funciones trigonom√©tricas, nos valdremos de estos ejes. Considerando un punto sobre la circunferencia de cordenadas (a,b) y un angulo $a$ en posicion normal asociada al punto, se define:



Sen ($a$) = $\frac{ordenada}{radio}$

cos ($a$) = $\frac{abscisa}{radio}$

tan ($a$) = $\frac{ordenada}{abscisa}$

Observaciones: 
* Los valores que toman las funciones son independientes del valor de la circunferencia $r$, por lo tanto para mayor simplicidad se puede considerar  $r$=1 
* la variable $a$ debe especificarse en unidades de radianes, puesto que las funciones son con variables reales

Se denomina funci√≥n seno, y se denota por f($x$)= Sen$x$, a la aplicaci√≥n de la raz√≥n trigonom√©trica seno a una variable independiente x expresada en radianes. La funci√≥n seno es peri√≥dica, acotada y continua, y su dominio de definici√≥n es el conjunto de todos los n√∫meros reales.

## La funci√≥n coseno

La funci√≥n coseno, que se denota por f ($x$) = Cos $x$, es la que resulta de aplicar la raz√≥n trigonom√©trica coseno a una variable independiente x expresada en radianes. Esta funci√≥n es peri√≥dica, acotada y continua, y existe para todo el conjunto de los n√∫meros reales.

## La funci√≥n tangente


Se define funci√≥n tangente de una variable num√©rica real a la que resulta de aplicar la raz√≥n trigonom√©trica tangente a los distintos valores de dicha variable. Esta funci√≥n se expresa gen√©ricamente como f($x$) = Tag $x$, siendo $x$ la variable independiente expresada en radianes.

Grafica de tangente La funci√≥n de la tangente es peri√≥dica de per√≠odo 180¬∫ (œÄ radianes).

##Propiedades de las funciones trigonom√©tricas

Como caracter√≠sticas importantes y distintivas de las funciones trigonom√©tricas pueden resaltarse las siguientes:

*Las funciones seno, coseno y tangente son de naturaleza peri√≥dica, de manera que el periodo de las funciones seno y coseno es 2$\pi$ y el de la funci√≥n tangente es $\pi$.


image.png




* Las funciones seno y coseno est√°n definidas para todo el conjunto de los n√∫meros reales. Ambas son funciones continuas (no as√≠ la funci√≥n tangente).

* Las funciones seno y coseno est√°n acotadas, ya que sus valores est√°n contenidos en el intervalo [-1,1]. La funci√≥n tangente no est√° acotada.

* Las funciones seno y tangente son sim√©tricas respecto al origen, ya que sen (-x) = -sen x; tag (-x)=-tag x. En cambio, la funci√≥n coseno es sim√©trica respecto al eje Y: cos (-x) = cos x.

Consideremos la circunferencia unitaria definida por la ecuaci√≥n 

$$ x^{2} + y^{2} = 1 $$


Entonces, las dos funciones que definen la semicircunferencia superior y la inferior son:

$$ \text{semicircunferencia superior}: y = \sqrt{1-x^{2}} $$


$$ \text{semicircunferencia inferior}: y = -\sqrt{1-x^{2}} $$


Como funciones de Python, estas semicircunferencias las podemos definir como:


```python
import numpy as np 
def semi_superior(x):
    return np.sqrt(1 - x**2)

import numpy as np 
def semi_inferior(x):
    return -semi_superior(x)
```

Como $1 - x^{2} \geq 0$, ya que esta es una funci√≥n de valores reales, entonces, el dominio de estas funciones es el intervalo cerrado de n√∫meros reales $[-1,1]$. As√≠, podemos obtener una representaci√≥n geom√©trica de la circunferencia unitaria, haciendo uso de estas funciones y utilizando matplotlib: 

```python
import matplotlib.pyplot as plt 

dominio = np.arange(-1,1, 0.00001)
rango1 = semi_superior(dominio)
plt.plot(dominio, rango1, 'g-')
plt.plot(dominio, -rango1, "g-")

plt.axis('equal')
```

Ahora, para trazar √°ngulos centrales positivos $t$, empecemos trazando el segmento radial $OA$
