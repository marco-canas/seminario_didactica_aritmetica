históricamente: como razones entre los lados de un triángulo rectángulo. Un

ángulo agudo, podemos considerar un triángulo rectángulo que tiene a θ como
uno de sus ángulos, como en la figura 1, donde el símbolo image.png  especifica el

##Importante ##

 Descripción verbal de la función seno

* tiene periodo:2π

* Es creciente en los intervalos de la forma:Primer y último cuadrante 

* Es decreciente en los intervalos de la forma:Segundo y tercer cuadrante 

* Su rango es acotado en el intervalo:[1.-1]



```python
import numpy as np
import matplotlib.pyplot as plt 

x = np.linspace(-2*np.pi, 4*np.pi, 200)
y = np.sin(x)
plt.plot(x,y)
plt.title(r'Función $y = \sin(x)$')

plt.grid(alpha = 0.6)

plt.show() 
```
